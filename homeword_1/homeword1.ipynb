```python


vocabulary = {
    "Size": [
        ("Big", "Lớn"),
        ("Small", "Nhỏ"),
        ("Huge", "Khổng lồ"),
        ("Enormous", "Khổng lồ"),
        ("Gigantic", "Cực lớn"),
        ("Giant", "To lớn"),
        ("Tiny", "Bé xíu"),
        ("Long", "Dài"),
        ("Short", "Ngắn"),
        ("Wide", "Rộng"),
        ("Large", "Rộng lớn"),
        ("Tall", "Cao"),
        ("Narrow", "Hẹp"),
        ("Slim", "Mỏng manh"),
        ("Thin", "Gầy, mỏng"),
        ("Thick", "Dày"),
        ("Vast", "Rộng lớn"),
        ("Overweight", "Thừa cân"),
        ("Broad", "Rộng"),
        ("Oversized", "Ngoại cỡ"),
        ("Chubby", "Mũm mĩm"),
        ("Massive", "To lớn, đồ sộ"),
        ("Endless", "Vô tận"),
        ("Boundless", "Không giới hạn"),
        ("Fat", "Mập mạp"),
        ("Miniature", "Nhỏ"),
        ("Skinny", "Gầy gò"),
        ("Undersized", "Nhỏ cỡ"),
        ("Underweight", "Nhẹ cân"),
        ("Heavy", "Nặng")
    ],
    "Color": [
        ("Pink", "Màu hồng"),
        ("Red", "Màu đỏ"),
        ("Green", "Màu xanh"),
        ("White", "Màu trắng"),
        ("Black", "Màu đen"),
        ("Orange", "Màu cam"),
        ("Blue", "Màu xanh da trời"),
        ("Purple", "Màu tím"),
        ("Brown", "Màu nâu"),
        ("Gray", "Màu xám"),
        ("Yellow", "Màu vàng")
    ],
    "Personality": [
        ("Brave", "Mạnh mẽ, dũng cảm"),
        ("Chatty", "Nhiều chuyện"),
        ("Clever", "Thông minh"),
        ("Cowardly", "Nhát gan"),
        ("Easy-going", "Dễ chịu"),
        ("Friendly", "Thân thiện"),
        ("Funny", "Hài hước"),
        ("Generous", "Rộng lượng"),
        ("Grumpy", "Dễ giận hờn"),
        ("Hard-working", "Chăm chỉ"),
        ("Honest", "Thật thà"),
        ("Kind", "Tử tế"),
        ("Lazy", "Lười biếng"),
        ("Loud", "Lớn tiếng"),
        ("Mean", "Ki bo, hà tiện"),
        ("Moody", "Thất thường"),
        ("Nasty", "Xấu tính"),
        ("Neat", "Gọn gàng"),
        ("Nice", "Tử tế và vui vẻ"),
        ("Polite", "Lịch sự"),
        ("Quiet", "Ít nói"),
        ("Rude", "Thô lỗ"),
        ("Selfish", "Ích kỉ"),
        ("Serious", "Nghiêm túc"),
        ("Shy", "Rụt rè, nhút nhát"),
        ("Silly", "Ngốc nghếch"),
        ("Smart", "Thông minh"),
        ("Stupid", "Ngu ngốc"),
        ("Wise", "Thông thái"),
        ("Optimistic", "Lạc quan"),
        ("Pessimistic", "Bi quan"),
        ("Humorous", "Khôi hài"),
        ("Patient", "Kiên nhẫn"),
        ("Careful", "Cẩn thận"),
        ("Energetic", "Tràn đầy năng lượng"),
        ("Trustworthy", "Đáng tin cậy"),
        ("Sensitive", "Nhạy cảm"),
        ("Independent", "Tự lập"),
        ("Persistent", "Kiên trì"),
        ("Confident", "Tự tin"),
        ("Sociable", "Hòa đồng"),
        ("Calm", "Điềm tĩnh"),
        ("Humble", "Khiêm tốn"),
        ("Frank", "Thẳng thắn"),
        ("Faithful", "Trung thành"),
        ("Aggressive", "Cộc cằn, nóng tính"),
        ("Stubborn", "Cứng đầu"),
        ("Arrogant", "Tự cao"),
        ("Indecisive", "Không chính kiến"),
        ("Thoughtless", "Không biết suy nghĩ"),
        ("Ignorant", "Thất học"),
        ("Irresponsible", "Thiếu trách nhiệm"),
        ("Jealous", "Ganh tỵ"),
        ("Unreliable", "Không đáng tin"),
        ("Careless", "Bất cẩn"),
        ("Intelligent", "Thông minh"),
        ("Idle", "Lười nhát"),
        ("Cruel", "Độc ác"),
        ("Ambitious", "Tham vọng"),
        ("Risky", "Mạo hiểm"),
        ("Caring", "Biết quan tâm người khác"),
        ("Active", "Năng động"),
        ("Passive", "Thụ động"),
        ("Obedient", "Vâng lời"),
        ("Mischievous", "Nghịch ngượm"),
        ("Curious", "Tò mò")
    ],
    "SHAPES": [
        ("Circular","Hình tròn"),
        ("Triangular",  "Hình tam giác"),
        ("Rectangular",  "Hình chữ nhật"),
        ("Vertical", "Thẳng đứng, hàng dọc"),
        ("Horizontal","Hàng ngang"),
        ("Curved","Hình cong"),
        ("Square","Hình vuông"),
        ("Straight","Đường thẳng")
    ],
    "TASTES": [
        ("Spicy", "Cay"),
        ("Sweet", "Ngọt"),
        ("Sour",  "Chua"),
        ("Bitter", "Đắng"),
        ("Salty","Mặn"),
        ("Delicious", "Ngon")
    ],
    "Others": [
        ("Important", "Quan trọng"),
        ("Beautiful", "Xinh đẹp"),
        ("Clean", "Sạch sẽ"),
        ("Dirty", "Dơ bẩn"),
        ("Peaceful", "Yên bình"),
        ("Noisy", "Ồn ào"),
        ("Expensive", "Đắt đỏ"),
        ("Cheap", "Rẻ tiền"),
        ("Rich", "Giàu có"),
        ("Poor", "Nghèo nàn"),
        ("Powerful", "Quyền lực"),
        ("Happy", "Vui vẻ, hạnh phúc"),
        ("Unhappy", "Không vui vẻ"),
        ("Lucky", "May mắn"),
        ("Unlucky", "Xui xẻo"),
        ("Interesting", "Thú vị"),
        ("Boring", "Buồn chán"),
        ("Wealthy", "Giàu có"),
        ("Healthy", "Khỏe mạnh"),
        ("Sick", "Bệnh"),
        ("Ill", "Bệnh"),
        ("Necessary", "Cần thiết"),
        ("Ugly", "Xấu xí"),
        ("Quick", "Nhanh"),
        ("Slow", "Chậm"),
        ("Old", "Già"),
        ("New", "Mới"),
        ("Full", "No, đầy"),
        ("Empty", "Trống rỗng"),
        ("Tired", "Mệt mỏi"),
        ("Sad", "Buồn"),
        ("Upset", "Đau buồn"),
        ("Disappointed", "Thất vọng"),
        ("Satisfied", "Hài lòng"),
        ("Excellent", "Xuất sắc"),
        ("Awful", "Tồi tệ, xấu xa"),
        ("Terrible", "Tồi tệ"),
        ("Horrible", "Kinh dị"),
        ("Successful", "Thành công"),
        ("Romantic", "Lãng mạn"),
        ("Wonderful", "Tuyệt vời"),
        ("Good", "Tốt"),
        ("Bad", "Xấu"),
        ("Strong", "Mạnh mẽ"),
        ("Popular", "Phổ biến"),
        ("Famous", "Nổi tiếng"),
        ("Effective", "Có hiệu quả"),
        ("Comfortable", "Thoải mái"),
        ("Difficult", "Khó"),
        ("Soft", "Mềm"),
        ("Hard", "Cứng, khó"),
        ("Tough", "Khô ráp"),
        ("Easy", "Dễ"),
        ("Ancient", "Cổ xưa"),
        ("Modern", "Hiện đại"),
        ("Transparent", "Trong suốt"),
        ("Wild", "Hoang dã"),
        ("Sorrowful", "Đau khổ"),
        ("Attractive", "Thu hút"),
        ("Safe", "An toàn"),
        ("Unsafe", "Không an toàn"),
        ("High", "Cao"),
        ("Low", "Thấp"),
        ("Disgusting", "Đáng kinh tởm"),
        ("Busy", "Bận rộn"),
        ("Respectful", "Tôn trọng"),
        ("Controversial", "Gây tranh cãi"),
        ("Persuasive", "Có tính thuyết phục"),
        ("Hungry", "Đói bụng"),
        ("Thirsty", "Khát nước"),
        ("Angry", "Giận"),
        ("Hot", "Nóng"),
        ("Cold", "Lạnh"),
        ("Grateful", "Biết ơn"),
        ("Thankful", "Cảm ơn"),
        ("Regretful", "Hối tiếc"),
        ("Beneficial", "Có lợi ích"),
        ("Useful", "Hữu ích"),
]

}
def similar(a, b):
    """So sánh độ tương đồng giữa hai chuỗi."""
    return SequenceMatcher(None, a, b).ratio()

def ai_vocabulary_learning():
    print("\n🌟 Chào mừng bạn đến với chương trình học từ vựng!")
    print("Dưới đây là các chủ đề bạn có thể học:")
    for idx, topic in enumerate(vocabulary.keys(), start=1):
        print(f"{idx}. {topic}")

    # Chọn chủ đề
    while True:
        try:
            topic_choice = int(input("\nNhập số thứ tự của chủ đề bạn muốn học: ")) - 1
            selected_topic = list(vocabulary.keys())[topic_choice]
            print(f"👉 Bạn đã chọn chủ đề: {selected_topic}")
            break
        except (ValueError, IndexError):
            print("❌ Lựa chọn không hợp lệ. Vui lòng nhập số thứ tự trong danh sách!")

    # Chọn giọng đọc
    while True:
        print("\nChọn giọng đọc bạn muốn nghe:")
        print("1. Đọc tiếng Anh")
        print("2. Đọc tiếng Việt")
        voice_choice = input("Nhập số 1 hoặc 2: ").strip()
        if voice_choice in ["1", "2"]:
            voice_lang = 'en' if voice_choice == "1" else 'vi'
            break
        else:
            print("❌ Lựa chọn không hợp lệ. Vui lòng thử lại!")

    # Lấy danh sách từ vựng theo chủ đề
    words_to_learn = vocabulary[selected_topic]
    random.shuffle(words_to_learn)

    incorrect_words = []  # Danh sách các từ sai

    print("\n✨ Bắt đầu bài học từ vựng!")

    while words_to_learn:
        for word, meaning in words_to_learn:
            # Tạo âm thanh dựa trên lựa chọn giọng đọc
            tts_text = word if voice_lang == 'en' else meaning
            tts = gTTS(text=tts_text, lang=voice_lang)
            tts.save("temp_audio.mp3")
            display(Audio("temp_audio.mp3", autoplay=True))  # Phát âm thanh tự động

            # Hiển thị hướng dẫn
            if voice_lang == 'en':
                print(f"Nghe từ và nhập từ tiếng Anh cùng nghĩa tiếng Việt:")
                user_word = input("Nhập từ tiếng Anh: ").strip()
                user_meaning = input("Nhập nghĩa tiếng Việt: ").strip()

                # Kiểm tra kết quả
                is_word_correct = user_word.lower() == word.lower()
                similarity = similar(user_meaning.lower(), meaning.lower())
                is_meaning_correct = similarity >= 0.7

                # Đưa ra kết quả
                if is_word_correct and is_meaning_correct:
                    print("\u2705 Đúng! Bạn đã trả lời chính xác cả từ tiếng Anh và nghĩa tiếng Việt.")
                else:
                    print("\u274c Sai rồi. Đây là đáp án đúng:")
                    print(f"- Từ tiếng Anh: {word}")
                    print(f"- Nghĩa tiếng Việt: {meaning}")
                    incorrect_words.append((word, meaning))
            else:
                print(f"Nghe và nhập từ tiếng Anh tương ứng với nghĩa: {meaning}")
                user_word = input("Nhập từ tiếng Anh: ").strip()

                # Kiểm tra từ tiếng Anh
                is_word_correct = user_word.lower() == word.lower()

                # Đưa ra kết quả
                if is_word_correct:
                    print("\u2705 Đúng! Bạn đã trả lời chính xác từ tiếng Anh.")
                else:
                    print("\u274c Sai rồi. Đây là đáp án đúng:")
                    print(f"- Từ tiếng Anh: {word}")
                    incorrect_words.append((word, meaning))

            print("-" * 50)
            os.remove("temp_audio.mp3")  # Xóa tệp sau khi sử dụng

        # Cập nhật danh sách từ sai
        words_to_learn = incorrect_words
        incorrect_words = []

    print("\n✨ Bạn đã trả lời chính xác tất cả các từ!\n")

    # Hỏi người dùng có muốn học tiếp không
    retry = input("\nBạn đã hoàn thành bài học! Bạn có muốn học lại không? (có/không): ").strip().lower()
    if retry in ("có", "co", "yes", "y"):
        ai_vocabulary_learning()
    else:
        print("Cảm ơn bạn đã học từ vựng! Chúc bạn học tốt.")

# Chạy chương trình học từ vựng
ai_vocabulary_learning()
```
